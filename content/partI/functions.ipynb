```python
# make this a function
import csv
def read_csv(file_name):
    newlines = []
    with open(file_name, 'r') as f:
        reader = csv.reader(f)
        for line in reader:
            newlines.append(line)
    return newlines
```

```python
data = read_csv('data/airports.csv')
```

```python
def get_attribute(data, attribute_name):
    header = data[0]
    idx = header.index(attribute_name)
    return [record[idx] for record in data[1:]]
        
```

```python
get_attribute(data, 'latitude')
```

```python
def to_float(attribute_list):
    return list(map(float, attribute_list))
```

```python
to_float(get_attribute(data, 'latitude'))
```

```python
lat = to_float(get_attribute(data, 'latitude'))
lon = to_float(get_attribute(data, 'longitude'))
```

```python

## distance between two points
## ignoring sphericality
```

```python
import math
def euclidean_distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    dx = x1 - x2
    dy = y1 - y2
    return math.sqrt(dx*dx + dy*dy)
```

```python
euclidean_distance((0,0), (1,1))
```

```python
# nearest neighbor
points = list(zip(lon, lat))
```

```python
min_lat = min(lat)
min_lon = min(lon)
max_lat = max(lat)
max_lon = max(lon)
max_d = euclidean_distance((min_lat, min_lon), (max_lat, max_lon))
```

```python
max_d
```

```python
# find nearest neighbor for point 0
p0 = points[0]
min_d = max_d
min_j = 0
for j, pj in enumerate(points):
    if j != 0:
        d = euclidean_distance(p0, pj)
        if d < min_d:
            min_d = d
            min_j = j
        
```

```python
# find the most distant neighbor for point 0
p0 = points[0]
max_d = 0.0
max_j = 0
for j, pj in enumerate(points):
    if j != 0:
        d = euclidean_distance(p0, pj)
        if d > max_d:
            max_d = d
            max_j = j
```
